```python
# Loading the dataset using pandas

import pandas as pd

df = pd.read_excel('Retail-Supply-Chain-Sales-Dataset.xlsx')

df.head(4)
```

```python
df.info()
```

```python
# Summary statistics using .describe()

df.describe(include='all')
```

```python
# Checking if missing date or null values are present in the dataset   

df.isnull().sum()
```

```python
df.columns.to_list()
```

```python
# Checking for duplicated 

df.duplicated().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'sub-category':'sub_category'})
```

```python
df.columns
```

```python
from sqlalchemy import create_engine

# Mysql Connection
username = "root"
password = "root"
host = "localhost"
port = "3306"
database = "retail_supply"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Write DataFrame to Mysql
table_name = "retail"  # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

# Read back sample
pd.read_sql("select * from retail limit 5;", engine)
```
